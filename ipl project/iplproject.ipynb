```python
#2)Which season had the most number of matches?(graph)

#Code:
plt.figure(figsize=(20,9))

sns.countplot(x='season',data=data2)

plt.show()
```

```python
plt.figure(figsize=(30,9))

sns.countplot(x='city',data=data2)

plt.show()
```

```python
 #4)graph
#Team won by max runs?
#And
#The best defending team?
#Code:
plt.figure(figsize=(5,12))

sns.swarmplot(x='win_by_runs',y='winner',data=data2)

plt.show()
```

```python
#5)Which Team had won by maximum runs? 
data2.iloc[data2['win_by_runs'].idxmax()]
```

```python
#6)graph-Which Team had won by maximum wickets? 
plt.figure(figsize=(5,12))

sns.swarmplot(x='win_by_wickets',y='winner',data=data2)

plt.show()
```

```python
#7)Which Team had won by minimum runs(closest margin) runs? 
data2.iloc[data2['win_by_runs'].idxmin()]
```

```python
#8)Which Team had won by minimum wickets?
data2.iloc[data2['win_by_wickets'].idxmin()]
```

```python
#10)Best Stadium to Bat first and Win**
#Best chasing venues?

#Code:

plt.figure(figsize=(5,12))

sns.swarmplot(x='win_by_wickets',y='venue',data=data2)

plt.show()

 #Eden Gardens and M Chinnaswamy Stadium are the best venues for chasing, the density of each attribute contributes in the swarmplot figure.
```
